```python
import random
import string
import numpy as np 
import pandas as pd
```

```python
random.seed(42)
np.random.seed(42)
```

```python
first_name=["Sonam","Komal","Gautam","poonam","Anshita","Samarth"]
last_name=["Saini","pundhir","kapoor","rajput","dawar","singh"]
```

```python
skills=["python","c","c++","MySql","Django"]
```

```python
interests=["AI","ML","WD","APP development"]
```

```python
colleges=["JMIT","JMIETI","ACE","KU","PU"]
```

```python
companies=["apple","google ","infosys","microsft"]
```

```python
roles=["student","student","student", "student","mentor","mentor"]
```

```python
#Unique Names
used_names=set()
def generate_unique_names():
    name = random.choice(first_name)+" "+random.choice(last_name)
    if name not in used_names:
        used_names.add(name)
        return name
```

```python
generate_unique_names()
```

```python
for i in range(10):
     print( generate_unique_names())
```

```python
# generate user 
def generate_user(user_id):
    role = random.choice(roles)
    name =generate_unique_names()
    selected_skills= random.sample(skills , random.randint(2,3))
    selected_interests= random.sample(interests , random.randint(2,4))
    if role=="Student":
        organization = random.choice(colleges)
        experience= random.randint(0,2)
    else:
        organization=random.choice(companies)
        experience= random.randint(3,20)
    bio= (f"{name} is a {role.lower()}interested in."
         f"{','.join(selected_interests[:2])}."
         f"Experienced in {','.join(selected_skills[:2])}.")
    return{
        "user_id":user_id,
        "name":name,
        "role":role,
        "organization":organization,
        "experience":skills,
        "intrests":interests,
        "bio":bio,
    }

    
```

```python
generate_user(1)
```

```python
users=[]
for i in range(10):
  users.append(generate_user(i))
print("Total  users:" , users)
```

```python
data=pd.DataFrame(users)
```

```python
data
```

```python
# Generate user connection
import networkx as nx
```

```python
print("Netwok Version:",nx.__version__)
```

```python
G= nx.Graph()
print(G)
```

```python
for _,row in data.iterrows():
    G.add_node(
        row["user_id"],
        name=row["name"],
        role=row["role"],
        organizaton=row["organization"]
    )
print("Total node:",G.number_of_nodes())
```

```python
print(G)
```

```python
#  generate some randim professional connections
all_users= data["user_id"].tolist()
connection= set()
for user in all_users:
    total_connections= random.randint(2,4)
    while total_connections >0:
        friend = random.choice(all_users)
        if friend!= user:
            edge = tuple(sorted((user, friend)))
            if edge not in connection:
                connection.add(edge)
                total_connections-=1
                




            
```

```python
G.add_edges_from(connection)
print("Total connections:",G.number_of_edges())
```

```python
#how to find he density of the graph
print("dense" ,round(nx.density(G),3))
```
