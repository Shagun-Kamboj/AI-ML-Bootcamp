```python
import numpy as np
import pandas as pd
np.random.seed(45)
n_student =10000
data ={
    "name" : np.random.choice(["dev","riya","shagun","poonam"]),
    "age" : np.random.randint(15,30,n_student),
    "city" : np.random.choice(["ynr","radaur","ambala"])
}
df=pd.DataFrame(data)
```

```python
import sqlite3 as sql
import pandas as pd
```

```python
connection = sql.connect('demo4.db')
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''CREATE TABLE data(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    city TEXT
);''')
```

```python
connection.commit()
```

```python
data1= list(df.itertuples(index=False, name= None))
cursor.executemany('''INSERT INTO data(name,age,city) VALUES (?,?,?)''',data1)

#cursor.execute('''INSERT INTO people(name,age,city)VALUES ('riya',19);''')   
```

```python
connection.commit()
```

```python
cursor.execute('''SELECT*FROM data''')
```

```python
results = cursor.fetchall()
```

```python
for record in results:
    print(record)
```

```python
df = pd.DataFrame(results)
```
