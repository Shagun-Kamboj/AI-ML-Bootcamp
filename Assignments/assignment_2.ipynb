1. What is the difference between a list and a tuple, beyond "one is mutable and one isn't"? 
    Why would you ever choose a tuple?</br>
Ans : Lists are mutable and consume more memory, while tuples are immutable, faster, and more memory-efficient. 
       Tuples can also be used as dictionary keys because they are hashable. Use tuples when data should not change.

2. Why is dictionary lookup O(1) on average? What data structure makes this possible internally?

Python dictionaries use a hash table internally.

When a key is inserted:
1. Python computes the hash value of the key.
2. The hash determines where the key-value pair is stored.
3. Looking up the key computes the same hash and directly accesses the location.

Therefore, dictionary lookup is **O(1) on average** because it does not need to search through all elements.

3. What does "hashable" mean, and which built-in types are NOT hashable?</br>
ANS : A hashable object has a fixed hash value that does not change during its lifetime. 
            Immutable types are usually hashable. Lists, dictionaries, and sets are not hashable.

5. Why is set membership testing faster than list membership testing?</br>
Ans: Sets use hash tables, giving O(1) average lookup time, 
            while lists require checking elements one by one, giving O(n) time complexity.

6. What happens in memory when you write b = a where a is a list? Draw or describe it.</br>
Ans: No new list is created. Both variables point to the same list object in memory.

**Bug 1 — Aliasing Trap**

This function is supposed to add a discount tag to a list of order</br>
WITHOUT modifying the caller's original list. It has a bug — find it.</br>
def apply_discount_tag(orders):</br>
 tagged = orders</br>
 tagged.append("DISCOUNT_APPLIED")</br>
 return tagged</br>
original_orders = ["order_1", "order_2"]</br>
new_orders = apply_discount_tag(original_orders)</br>
print("Original:", original_orders) # Did this change unexpectedly?</br>
print("New:", new_orders)</br>
**Fix ↓** 

```python
def apply_discount_tag(orders):
    tagged = orders.copy()   # new copy
    tagged.append("DISCOUNT_APPLIED")
    return tagged

original_orders = ["order_1", "order_2"]
new_orders = apply_discount_tag(original_orders)

print("Original:", original_orders)
print("New:", new_orders)
```
