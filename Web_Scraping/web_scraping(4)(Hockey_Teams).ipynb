```python
import requests
import pandas as pd
```

```python
def fetch_hockey_teams():

    url = "https://www.thesportsdb.com/api/v1/json/3/search_all_teams.php?l=NHL"

    try:
        response = requests.get(url)
        response.raise_for_status()

        data = response.json()

        teams = []

        if data.get("teams"):

            for team in data["teams"]:

                teams.append({
                    "Team Name": team.get("strTeam"),
                    "Country": team.get("strCountry"),
                    "Formed Year": team.get("intFormedYear"),
                    "Stadium": team.get("strStadium"),
                    "Badge": team.get("strBadge")
                })

            return teams

        else:
            print("No teams found.")
            return []

    except Exception as e:
        print("Error:", e)
        return []
```

```python
teams = fetch_hockey_teams()
```

```python
print("Total Teams:", len(teams))
```

```python
df = pd.DataFrame(teams)

df
```

```python
df.to_csv("hockey_teams.csv", index=False)

print("CSV Saved Successfully")
```
