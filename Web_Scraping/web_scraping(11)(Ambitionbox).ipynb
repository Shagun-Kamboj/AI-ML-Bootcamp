```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_ambitionbox():
    url = "https://www.ambitionbox.com/list-of-companies"

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    companies = []

    cards = soup.find_all("div", class_="companyCardWrapper")

    for card in cards[:20]:

        try:
            name = card.find("h2").get_text(strip=True)
        except:
            name = "N/A"

        try:
            rating = card.find("span", class_="companyCardWrapper__companyRatingValue").get_text(strip=True)
        except:
            rating = "N/A"

        try:
            reviews = card.find("span", class_="companyCardWrapper__ActionCount").get_text(strip=True)
        except:
            reviews = "N/A"

        companies.append({
            "name": name,
            "rating": rating,
            "reviews": reviews
        })

    return companies
```

```python
companies = scrape_ambitionbox()
print(companies)
```
