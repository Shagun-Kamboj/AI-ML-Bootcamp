```python
import sqlite3 as sql
import pandas as pd
```

```python
# connect to a database[Create if it does not exist]
connection = sql.connect('demo1.db')
```

```python
#create a cursor object using the connection
cursor = connection.cursor()
# cursor helps you to traverse over the records(rows) in a db
```

```python
cursor.execute('''CREATE TABLE people(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    city TEXT
);''')
```

```python
connection.commit()
```

```python
cursor.execute('''INSERT INTO people(name,age,city) VALUES ('SHAGUN',19,'YNR');''')
cursor.execute('''INSERT INTO people(name,age,city)VALUES ('riya',19,'ynr');''')                                                    
```

```python
connection.commit()
```

```python
cursor.execute('''SELECT*FROM people''')
```

```python
results = cursor.fetchall()
```

```python
for record in results:
    print(record)
```

```python
df = pd.DataFrame(results)
```
